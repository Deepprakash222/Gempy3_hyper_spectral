```python
import numpy as np 
import matplotlib.pyplot as plt 
import json
```

```python
# Path to the JSON file
file_path = './posterior_samples.json'

# Open and read the JSON file
with open(file_path, 'r') as file:
    data = json.load(file)

# Now `data` is a Python dictionary containing the parsed JSON
print(data)
```

```python
cov_eigen_values_1 = np.array(data['cov_eigen_values_1'])
cov_eigen_values_2 = np.array(data['cov_eigen_values_2'])
cov_eigen_values_3 = np.array(data['cov_eigen_values_3'])
cov_eigen_values_4 = np.array(data['cov_eigen_values_4'])
cov_eigen_values_5 = np.array(data['cov_eigen_values_5'])
cov_eigen_values_6 = np.array(data['cov_eigen_values_6'])
mean_data_1        = np.array(data['mean_data_1'])
mean_data_2        = np.array(data['mean_data_2'])
mean_data_3        = np.array(data['mean_data_3'])
mean_data_4        = np.array(data['mean_data_4'])
mean_data_5        = np.array(data['mean_data_5'])
mean_data_6        = np.array(data['mean_data_6'])
eigen_vectors      = np.array(data["eigen_vectors"])
mu_1               = np.array(data["mu_1"])
mu_2               = np.array(data["mu_2"])
mu_3               = np.array(data["mu_3"])
mu_4               = np.array(data["mu_4"])
```

```python
mean_mu_1 = np.mean(mu_1)
std_mu_1 = np.std(mu_1)
mean_mu_2 = np.mean(mu_2)
std_mu_2 = np.std(mu_2)
mean_mu_3 = np.mean(mu_3)
std_mu_3 = np.std(mu_3)
mean_mu_4 = np.mean(mu_4)
std_mu_4 = np.std(mu_4)
mean_mu_1, std_mu_1, mean_mu_2, std_mu_2, mean_mu_3, std_mu_3,mean_mu_4, std_mu_4
```

```python
mean_mean_1 = np.mean(mean_data_1, axis=0)
mean_std_1 = np.std(mean_data_1, axis=0)
mean_mean_2 = np.mean(mean_data_2, axis=0)
mean_std_2= np.std(mean_data_2, axis=0)
mean_mean_3 = np.mean(mean_data_3, axis=0)
mean_std_3 = np.std(mean_data_3, axis=0)
mean_mean_4 = np.mean(mean_data_4, axis=0)
mean_std_4 = np.std(mean_data_4, axis=0)
mean_mean_5 = np.mean(mean_data_5, axis=0)
mean_std_5 = np.std(mean_data_5, axis=0)
mean_mean_6 = np.mean(mean_data_6, axis=0)
mean_std_6 = np.std(mean_data_6, axis=0)
```

```python
cov_1 = []
for i in range(cov_eigen_values_1.shape[0]):
    cov = eigen_vectors[0] @ np.diag(cov_eigen_values_1[i,:])**2 @ eigen_vectors[0].T
    cov_1.append(cov)
cov_1 = np.array(cov_1)
mean_cov_1 = np.mean(cov_1,axis=0)
std_cov_1 = np.std(cov_1,axis=0)
mean_cov_1.shape, std_cov_1.shape
```

```python
cov_4 = []
for i in range(cov_eigen_values_2.shape[0]):
    cov = eigen_vectors[3] @ np.diag(cov_eigen_values_4[i,:])**2 @ eigen_vectors[3].T
    cov_4.append(cov)
cov_4 = np.array(cov_4)
mean_cov_4 = np.mean(cov_4,axis=0)
std_cov_4 = np.std(cov_4,axis=0)
mean_cov_4.shape, std_cov_4.shape
```

```python
cov_5 = []
for i in range(cov_eigen_values_2.shape[0]):
    cov = eigen_vectors[4] @ np.diag(cov_eigen_values_5[i,:])**2 @ eigen_vectors[4].T
    cov_5.append(cov)
cov_5 = np.array(cov_5)
mean_cov_5 = np.mean(cov_5,axis=0)
std_cov_5 = np.std(cov_5,axis=0)
mean_cov_5.shape, std_cov_5.shape
```

```python
cov_6 = []
for i in range(cov_eigen_values_2.shape[0]):
    cov = eigen_vectors[5] @ np.diag(cov_eigen_values_6[i,:])**2 @ eigen_vectors[5].T
    cov_6.append(cov)
cov_6 = np.array(cov_6)
mean_cov_6 = np.mean(cov_6,axis=0)
std_cov_6 = np.std(cov_6,axis=0)
mean_cov_6.shape, std_cov_6.shape
```

```python
means =[]
means.append(mean_mean_1)
means.append(mean_mean_2)
means.append(mean_mean_3)
means.append(mean_mean_4)
means.append(mean_mean_5)
means.append(mean_mean_6)

covariances=[]
covariances.append(mean_cov_1)
covariances.append(mean_cov_2)
covariances.append(mean_cov_3)
covariances.append(mean_cov_4)
covariances.append(mean_cov_5)
covariances.append(mean_cov_6)
```

```python
from scipy.stats import multivariate_normal
```

```python


K = 6 # Number of cluster
D = len(means[0])  # 8 Dimesion of feature space

n =3 # how far in terms of std you want to plot

# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=D, ncols=D, figsize=(15, 15))
label_to_color = { 0: 'red', 1: 'blue', 2: 'green', 3: 'yellow', 4: 'orange', 5: 'purple'}

for i in range(K):
    # Create different plot of each cluster
    for j in range(D):
        for k in range(D):
            if j==k:
                mean_ = np.array([means[i][j],means[i][k]])
                cov_  = np.diag([covariances[i][j][j], covariances[i][k][k]])
                
                # Create a grid of points
                x, y = np.meshgrid(np.linspace(mean_[0] -n * np.sqrt(cov_[0][0]), mean_[0] +n * np.sqrt(cov_[0][0]), 1000), np.linspace(mean_[1]-n * np.sqrt(cov_[1][1]), mean_[1]+ n * np.sqrt(cov_[1][1]), 1000))
                pos = np.dstack((x, y))
                # Create the multivariate normal distribution
                rv = multivariate_normal(mean_, cov_)
                # Compute the probability density function (PDF) at the grid points
                pdf = rv.pdf(pos)
                axes[j][k].contour(x, y, pdf, levels=1,colors=label_to_color[i], linewidths=2)
                
                
            else:
                mean_ = np.array([means[i][j],means[i][k]])
                cov_  = np.array([[covariances[i][j][j],covariances[i][j][k]],[covariances[i][k][j], covariances[i][k][k]]])
                
                # Create a grid of points
                x, y = np.meshgrid(np.linspace(mean_[0] -n * np.sqrt(cov_[0][0]), mean_[0] +n * np.sqrt(cov_[0][0]), 1000), np.linspace(mean_[1]-n * np.sqrt(cov_[1][1]), mean_[1]+ n * np.sqrt(cov_[1][1]), 1000))
                pos = np.dstack((x, y))
                # Create the multivariate normal distribution
                rv = multivariate_normal(mean_, cov_)
                # Compute the probability density function (PDF) at the grid points
                pdf = rv.pdf(pos)
                axes[j][k].contour(x, y, pdf, levels=1, colors=label_to_color[i], linewidths=2)
                
            #
                
            if j==0:
                axes[j][k].set_title(f'feature_{k+1}')
            if k==0:
                axes[j][k].set_ylabel(f'feature_{j+1}')

# Add a legend
lines = [] 
labels = [] 
  
for ax in fig.axes: 
    Line, Label = ax.get_legend_handles_labels() 
    lines.extend(Line) 
    labels.extend(Label)

plt.show() 
```
